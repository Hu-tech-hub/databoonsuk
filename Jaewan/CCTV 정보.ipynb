```python
import pandas as pd
import folium
import json
from shapely.geometry import shape
from collections import defaultdict

# 1. 데이터 불러오기
df = pd.read_csv('서울_경찰청_범죄 발생 지역별 통계.csv')
with open('HangJeongDong_ver20241001.geojson', encoding='utf-8') as f:
    geo = json.load(f)

# 2. GeoJSON 처리: 구 단위 중심 좌표 계산
gu_coords = defaultdict(list)

for feature in geo['features']:
    props = feature['properties']
    gu_name = props['sggnm']  # '종로구', '중구' 등
    geom = shape(feature['geometry'])
    lon, lat = geom.centroid.xy
    gu_coords[gu_name].append((lat[0], lon[0]))

# 구별 평균 중심좌표
center_dict = {
    gu: (
        sum(lat for lat, _ in pts) / len(pts),
        sum(lon for _, lon in pts) / len(pts)
    )
    for gu, pts in gu_coords.items()
}

# 3. CSV 전처리: Wide → Long 변환
df_long = df.melt(id_vars=['범죄대분류', '범죄중분류'], var_name='자치구', value_name='발생건수')
df_long = df_long.dropna()
df_long = df_long[df_long['발생건수'] > 0]

# 4. 자치구별 Top3 범죄 유형 및 총합 계산
df_grouped = df_long.groupby(['자치구', '범죄대분류']).agg({'발생건수': 'sum'}).reset_index()

crime_summary = df_grouped.groupby('자치구').apply(
    lambda x: pd.Series({
        '총범죄': x['발생건수'].sum(),
        'top범죄': ', '.join(x.sort_values('발생건수', ascending=False).head(3)['범죄대분류'].tolist())
    })
).reset_index()

# 5. Folium 지도 생성
m = folium.Map(location=[37.5665, 126.9780], zoom_start=11)

# 6. 원형 마커 추가
for _, row in crime_summary.iterrows():
    gu = row['자치구']
    if gu in center_dict:
        lat, lon = center_dict[gu]
        tooltip = f"<b>{gu}</b><br>총범죄: {row['총범죄']}<br>Top3: {row['top범죄']}"
        folium.CircleMarker(
            location=[lat, lon],
            radius=row['총범죄'] ** 0.5 * 0.5,
            color='crimson',
            fill=True,
            fill_opacity=0.6,
            tooltip=tooltip
        ).add_to(m)
    else:
        print(f"❗ {gu}는 center_dict에 없음")

# 7. 출력 및 저장
m.save('crime_map.html')
m  # 주피터랩 내에서 지도 보기
```
