```python
import pandas as pd
import folium
import json
from shapely.geometry import shape
from collections import defaultdict

# 1.  
df = pd.read_csv('__   .csv')
with open('HangJeongDong_ver20241001.geojson', encoding='utf-8') as f:
    geo = json.load(f)

# 2. GeoJSON :     
gu_coords = defaultdict(list)

for feature in geo['features']:
    props = feature['properties']
    gu_name = props['sggnm']  # '', '' 
    geom = shape(feature['geometry'])
    lon, lat = geom.centroid.xy
    gu_coords[gu_name].append((lat[0], lon[0]))

#   
center_dict = {
    gu: (
        sum(lat for lat, _ in pts) / len(pts),
        sum(lon for _, lon in pts) / len(pts)
    )
    for gu, pts in gu_coords.items()
}

# 3. CSV : Wide  Long 
df_long = df.melt(id_vars=['', ''], var_name='', value_name='')
df_long = df_long.dropna()
df_long = df_long[df_long[''] > 0]

# 4.  Top3     
df_grouped = df_long.groupby(['', '']).agg({'': 'sum'}).reset_index()

crime_summary = df_grouped.groupby('').apply(
    lambda x: pd.Series({
        '': x[''].sum(),
        'top': ', '.join(x.sort_values('', ascending=False).head(3)[''].tolist())
    })
).reset_index()

# 5. Folium  
m = folium.Map(location=[37.5665, 126.9780], zoom_start=11)

# 6.   
for _, row in crime_summary.iterrows():
    gu = row['']
    if gu in center_dict:
        lat, lon = center_dict[gu]
        tooltip = f"<b>{gu}</b><br>: {row['']}<br>Top3: {row['top']}"
        folium.CircleMarker(
            location=[lat, lon],
            radius=row[''] ** 0.5 * 0.5,
            color='crimson',
            fill=True,
            fill_opacity=0.6,
            tooltip=tooltip
        ).add_to(m)
    else:
        print(f" {gu} center_dict ")

# 7.   
m.save('crime_map.html')
m  #    
```

```python
total_counts = grouped.sum(axis=1)

#  5  
top5_categories = total_counts.sort_values(ascending=False).head(5).index
top5_grouped = grouped.loc[top5_categories]

# 
top5_grouped.T.plot(kind='bar', figsize=(14, 6))
plt.title('     ( 5)')
plt.xlabel('')
plt.ylabel('  ')
plt.legend(title='', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
top5_ratio = top5_grouped.div(top5_grouped.sum(axis=1), axis=0)

#     
top5_ratio.T.plot(kind='bar', stacked=True, figsize=(14, 6))
plt.title('    ( 5)')
plt.xlabel('')
plt.ylabel('')
plt.legend(title='', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
